```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("weather.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df["temparature"].fillna(df["temparature"].mean(),inplace=True)
df
```

```python
df["winspeed"].fillna(df["winspeed"].mean(),inplace=True)
df
```

```python
df["event"].mode()
```

```python
df["event"].fillna("fullair",inplace=True)
df
```

```python
df.info()
```

```python
df.drop("s.no",axis=1,inplace=True)
df
```

```python
df.drop_duplicates(["day"],keep="first",inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['day']=le.fit_transform(df['day'])
df['event']=le.fit_transform(df['event'])
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import Normalizer
s=StandardScaler()
m=MinMaxScaler()
r=RobustScaler()
n=Normalizer()
```

```python
s1=s.fit_transform(df) 
m1=m.fit_transform(df)
r1=r.fit_transform(df)
n1=n.fit_transform(df)
```

```python
plt.plot(df,marker="o")
plt.title("before scaling")
plt.legend(["Day","Temparature","windspeed","sales"])
plt.show()
```

```python
plt.plot(s1,marker="o")
plt.title("After Applying Standard  scaling")
plt.legend(["Day","Temparature","windspeed","sales"])
plt.show()
```

```python
plt.plot(m1,marker="o")
plt.title("After Applying minmax  scaling")
plt.legend(["Day","Temparature","windspeed","sales"])
plt.show()
```

```python
plt.plot(r1,marker="o")
plt.title("After Applying Robust  scaling")
plt.legend(["Day","Temparature","windspeed","sales"])
plt.show()
```

```python
plt.plot(n1,marker="o")
plt.title("After Applying  Normalizer  scaling")
plt.legend(["Day","Temparature","windspeed","sales"])
plt.show()
```

```python
plt.plot(df["temparature"],df["event"],marker="*")
plt.title("Temparature vs Event")
plt.xlabel("Temparature")
plt.ylabel("Event")
plt.show()
```

```python
plt.subplots(1,3)
plt.subplot(1,3,1)
plt.plot(df["day"],df["event"],marker="*")
plt.title("Day vs Event")
plt.xlabel("Day")
plt.ylabel("Event")
plt.subplot(1,3,2)
plt.plot(df["temparature"],df["event"],marker="*")
plt.title("Temparature vs Event")
plt.xlabel("Temparature")
plt.ylabel("Event")
plt.subplot(1,3,3)
plt.plot(df["winspeed"],df["event"],marker="*")
plt.title("Windspeed vs Event")
plt.xlabel("Windspeed")
plt.ylabel("Event")
plt.subplots_adjust(left=0,bottom=0.1, right=2, top=0.9,wspace=0.4,hspace=1)
plt.show()
```
